```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def tanh(x):
    return np.tanh(x)

def relu(x):
    return np.maximum(0, x)

def softmax(x):
    #exp_x = np.exp(x - np.max(x))
    #return exp_x / np.sum(exp_x, axis=0)
    return np.exp(x)/(np.sum(np.exp(x),axis=0))
    
x = np.linspace(-10, 10, 1000)

plt.plot(x,x)
plt.figure(figsize=(10, 6))
plt.subplot(221)
plt.plot(x, sigmoid(x))
plt.title('Sigmoid')
plt.subplot(222)
plt.plot(x, tanh(x))
plt.title('Tanh')
plt.subplot(223)
plt.plot(x, relu(x))
plt.title('ReLU')
plt.subplot(224)
y = softmax(np.array([1, 2, 3, 4]))
plt.plot(x, softmax(x))
plt.title('Softmax')
plt.show()
```
