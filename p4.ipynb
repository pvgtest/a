```python
import numpy as np 
import matplotlib.pyplot as plt

x=np.array([[-1,-1],[-1,1],[1,-1],[1,1],])
y=np.array([-1,-1,-1,1])

b=0
w=np.zeros(x.shape[1])
l=0.3

for _ in range(20):
    for i in range(x.shape[0]):
        p=np.sign(np.dot(x[i],w)+b)
        if p != y[i]:
            w+=l*y[i]*x[i]
            b+=l*y[i]

slope=-w[0]/w[1]
intercept=-b/w[1]

x_max,x_min=plt.xlim()
plt.scatter(x[:,0],x[:,1])
plt.plot([x_min,x_max],[x_min*slope*intercept,x_max*slope*intercept])

plt.show()

```
